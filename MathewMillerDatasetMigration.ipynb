## My notebook containing the original dataset

```python
# Load Modules

# Python package for dataframe generation and manipulation
import pandas as pd
# Python package for third party operating systems
import os
```

```python
# Read File

honeybee_csv1 = pd.read_csv(os.path.join('Resources', 'Data', 'vHoneyNeonic_v02.csv'))
honeybee_csv1.head()
```

```python
# Read File

honeybee_csv2 = pd.read_csv(os.path.join('Resources', 'Data', 'vHoneyNeonic_v03.csv'))
honeybee_csv2
```

```python
# Drop Null Values

honeybee_csv2 = honeybee_csv2.dropna(how='any',axis=0) 
honeybee_csv2
```

```python
# List Column Names

reorganize_columns = honeybee_csv2.columns.tolist()
reorganize_columns
```

```python
# Reorganize Column Arrangement

honeybee_csv2 = honeybee_csv2[['state', 'StateName', 'Region', 'FIPS', 'numcol', 'totalprod', 'yieldpercol', 'priceperlb', 'prodvalue', 'stocks', 'year', 'nCLOTHIANIDIN', 'nIMIDACLOPRID', 'nTHIAMETHOXAM', 'nACETAMIPRID', 'nTHIACLOPRID', 'nAllNeonic']]
honeybee_csv2
```

```python
# Rename Columns

honeybee_csv2 = honeybee_csv2.rename(columns={'state': 'State', 'numcol': 'ColonyCount', 'totalprod': 'TotalProduction',
                                             'yieldpercol': 'YieldPerColony', 'priceperlb': 'PricePerLB', 'prodvalue': 'ProductionValue',
                                             'stocks': 'Stocks', 'year': 'YearRecorded', 'nCLOTHIANIDIN': 'Clothianidin',
                                             'nIMIDACLOPRID': 'Imidacloprid', 'nTHIAMETHOXAM': 'Thiamethoxam', 'nACETAMIPRID': 'Acetamiprid',
                                             'nTHIACLOPRID': 'Thiacloprid', 'nAllNeonic': 'CombinedNeonic'})
honeybee_csv2
```

## Column Appendix:
* State = State name abbreviation
* StateName = State name in full
* Region = Classification of States per US-Census
* FIPS = Federal Information Processing System code
* ColonyCount = The States total number of Honeybee colonys 
* TotalProduction = The States total amount of honey produced
* YieldPerColony = Total production devided by the colony count (TotalProduction / ColonyCount)
* PricePerLB = Based on sales calculated in US dollars
* ProductionValue = Total production multiplied by price per pound in US dollars (TotalProduction x PricePerLB)
* Stocks = Honey reserves held by the honey farmers per pound
* YearRecorded = The year that all of the measurements were recorded for a given row
* Clothianidin = The amount of Clothianidin droplets sprayed directly onto plants in kg
* Imidacioprid = The amount of Imidacioprid droplets sprayed directly onto plants in kg
* Thiamethoxam = The amount of Thiamethoxam droplets sprayed directly onto plants in kg
* Acetamiprid = The amount of Acetamiprid droplets sprayed directly onto plants in kg
* Thiacioprid = The amount of Thiacioprid droplets sprayed directly onto plants in kg

```python
# Export to CSV

honeybee_csv2.to_csv("Resources/Output/CleanedHoneybeeDataset.csv", index=False, header=True)
```
