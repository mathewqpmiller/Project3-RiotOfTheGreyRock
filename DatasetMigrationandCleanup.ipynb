## Column Appendix:
* State = State name abbreviation
* StateName = State name in full
* Region = Classification of States per US-Census
* FIPS = Federal Information Processing System code
* ColonyCount = The States total number of Honeybee colonys 
* TotalProduction = The States total amount of honey produced
* YieldPerColony = Total production devided by the colony count (TotalProduction / ColonyCount)
* PricePerLB = Based on sales calculated in US dollars
* ProductionValue = Total production multiplied by price per pound in US dollars (TotalProduction x PricePerLB)
* Stocks = Honey reserves held by the honey farmers per pound
* YearRecorded = The year that all of the measurements were recorded for a given row
* Clothianidin = The amount of Clothianidin droplets sprayed directly onto plants in kg
* Imidacioprid = The amount of Imidacioprid droplets sprayed directly onto plants in kg
* Thiamethoxam = The amount of Thiamethoxam droplets sprayed directly onto plants in kg
* Acetamiprid = The amount of Acetamiprid droplets sprayed directly onto plants in kg
* Thiacioprid = The amount of Thiacioprid droplets sprayed directly onto plants in kg
* ColonyGrowth = Different in colony size in each state over each year of recorded data
* GrowthOutcome = Binary notation of colony growth (ColonyGrowth <= 0 returns a '0', ColonyGrowth > 0 returns a '1')
